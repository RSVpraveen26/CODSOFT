```python
%%writefile app.py
import streamlit as st
from crewai import Agent, Task, Crew, Process
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_community.utilities import SerpAPIWrapper  # Fixed path!
from langchain.tools import Tool
import os
from datetime import datetime
import io
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet

st.set_page_config(page_title="WanderWise AI", page_icon="🌍")
st.title("🌿 **WanderWise AI** – *Eco Travel Concierge*")
st.markdown("**Colab-la run pannuren! VS Code illa? No problem!**")

# API Keys
gemini_key = st.text_input("Gemini API Key (Free)", type="password", help="Get at [aistudio.google.com](https://aistudio.google.com)")
serpapi_key = st.text_input("SerpAPI Key (Free 100 searches)", type="password", help="Get at [serpapi.com](https://serpapi.com)")

if not gemini_key or not serpapi_key:
    st.warning("**API keys podu da!** Gemini: [aistudio.google.com](https://aistudio.google.com) | SerpAPI: [serpapi.com](https://serpapi.com)")
    st.stop()

os.environ["GEMINI_API_KEY"] = gemini_key
os.environ["SERPAPI_KEY"] = serpapi_key

# LLM & Tools
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0.3)
search = SerpAPIWrapper()

def get_weather(city):
    return search.run(f"7 day weather forecast {city}")

weather_tool = Tool(
    name="WeatherTool",
    func=get_weather,
    description="Get 7-day weather forecast for a city"
)

# Agents
flight_agent = Agent(
    role="Eco Flight Scout",
    goal="Find low-carbon, budget flights",
    backstory="Sustainable aviation expert",
    tools=[search],
    llm=llm,
    verbose=False,
    memory=True
)

hotel_agent = Agent(
    role="Green Hotel Hunter",
    goal="Eco-certified hotels",
    backstory="Green hospitality lover",
    tools=[search],
    llm=llm,
    verbose=False,
    memory=True
)

food_agent = Agent(
    role="Local Food Guide",
    goal="Authentic, dietary meals",
    backstory="Street food & veg expert",
    tools=[search],
    llm=llm,
    verbose=False,
    memory=True
)

planner = Agent(
    role="Master Itinerary Planner",
    goal="Weather-aware full plans",
    backstory="Journey architect",
    tools=[weather_tool],
    llm=llm,
    verbose=False,
    memory=True
)

# UI
col1, col2 = st.columns(2)
with col1:
    st.subheader("Trip Details")
    city = st.text_input("Destination City", "Goa, India")
    budget = st.number_input("Total Budget (USD)", min_value=300, max_value=5000, value=800)
    start = st.date_input("Start Date", datetime(2025, 12, 20))

with col2:
    end = st.date_input("End Date", datetime(2025, 12, 24))
    prefs = st.text_area("Preferences (e.g., veg, beach, adventure)", "vegetarian, beaches, no crowds", height=80)

if st.button("🧳 Generate Eco Itinerary!", use_container_width=True):
    with st.spinner("🤖 Agents planning your trip..."):
        days = (end - start).days + 1

        tasks = [
            Task(
                description=f"Find eco-friendly flights to {city} from {start.strftime('%Y-%m-%d')} to {end.strftime('%Y-%m-%d')}, max ${budget//3}. Output JSON: airline, price, CO2.",
                agent=flight_agent,
                expected_output="JSON flight details"
            ),
            Task(
                description=f"Recommend {days}-night eco hotel in {city}, max ${budget//2}. Output JSON: name, price, rating.",
                agent=hotel_agent,
                expected_output="JSON hotel info"
            ),
            Task(
                description=f"Plan {days*3} meals in {city} for {prefs}, avg ${budget//(days*15)}. Output JSON day-wise.",
                agent=food_agent,
                expected_output="JSON meal plan"
            ),
            Task(
                description=f"Build {days}-day itinerary for {city}: ${budget}, {prefs}, weather-aware. Use previous outputs.",
                agent=planner,
                expected_output="Markdown daily plan"
            )
        ]

        crew = Crew(
            agents=[flight_agent, hotel_agent, food_agent, planner],
            tasks=tasks,
            process=Process.sequential,
            verbose=1,
            memory=True
        )

        result = crew.kickoff()

    st.success("✨ **Itinerary Ready!**")
    st.markdown("---")
    st.markdown(result)

    # PDF Export
    buffer = io.BytesIO()
    doc = SimpleDocTemplate(buffer, pagesize=letter)
    styles = getSampleStyleSheet()
    story = [
        Paragraph(f"<b>🌿 WanderWise AI – {city} Eco Trip</b>", styles['Title']),
        Spacer(1, 12),
        Paragraph(f"<i>Dates: {start.strftime('%b %d')}–{end.strftime('%b %d')} | Budget: ${budget} | Prefs: {prefs}</i>", styles['Italic']),
        Spacer(1, 12)
    ]
    for line in result.split('\n'):
        if line.strip():
            story.append(Paragraph(line.replace('**', '<b>').replace('*', '<i>'), styles['Normal']))
            story.append(Spacer(1, 6))
    doc.build(story)

    st.download_button(
        label="📥 Download PDF Itinerary",
        data=buffer.getvalue(),
        file_name=f"WanderWise_{city.replace(' ', '_')}_{start.strftime('%b%d')}.pdf",
        mime="application/pdf"
    )

st.caption("**Kaggle AI Agents Capstone 2025 | Concierge Track**")
```
